```python
import pandas as pd
import scattertext as st
import spacy
from pprint import pprint
```

```python
convention_df = st.SampleCorpora.ConventionData2012.get_data()
convention_df.head()
```

```python
ted_talks = pd.read_json('interactive-visual/ted-talks.json', orient='records')
ted_talks.head()
```

```python
nlp = spacy.load('en')
corpus = st.CorpusFromPandas(ted_talks, category_col='event', text_col='transcript', nlp=nlp).build()
```

```python
print(list(corpus.get_scaled_f_scores_vs_background().index[:10]))
```

```python
html = st.produce_scattertext_explorer(corpus.get_unigram_corpus().compact(st.ClassPercentageCompactor(term_count=2, term_ranker=st.OncePerDocFrequencyRanker)),
            category='TED2018',
            category_name='TED2018',
            not_category_name='Other',
            width_in_pixels=1000)
#             ,metadata=ted_talks['speaker'])
```

```python
open("TED2018-Visualization.html", 'wb').write(html.encode('utf-8'))
```
